# Iris dataset

Este famoso conjunto de datos corresopnde a un análisis multivariado que el biólogo Ronald Fisher utilizó en su artículo de 1936 *The use of multiple measurements in taxonomic problems* para ejemplificar un análisis discriminante lineal.

Los datos consisten en 4 mediciones para tres especies de flor Iris (Iris setosa, Iris virginica and Iris versicolor). Las 4 mediciones corresponden a la longitud y altura de los sépalos y de los pétalos de cada especimen.

# Actividad

```
# Carga los datos y muestra los primeros renglones
import numpy as np
import pandas as pd

df = pd.read_csv('data/iris.csv')

df.head()
```

```
# Crea una tabla resumen con los estadísticos generales de las variables numéricas
df.describe()
```

```
# ¿Cuántas observaciones hay de cada tipo de flor?
df["Species"].value_counts()
```

*50 de cada una*

```
# ¿Cómo se comparan los pétalos y sépalos para las distintas especies de flor?
# Calcula los estadísticos de tendencia central y dispersión para cada especie.
df.groupby("Species").mean()
```

**Promedio**

```
df.groupby("Species").std()
```

```
# ¿Qué relación hay entre las 4 variables? Obtén la matriz de correlación y escribe tus conclusiones.
df.corr()
```

*Hay correlación fuerte positiva entre la longitud del sépalo y la longitud y ancho del pétalo. También hay correlación muy fuerte positiva entre el ancho del pétalo y su longitud. Esto quiere decir que hay relación entre estas variables estudiadas y mencionadas.*
